<font color='yellow'>This dataset has 414 sample data . We will create a ML model to predict the price per unit area of a house based on -

1.   Age of the house
2.   Distance from nearest MRT satiation
3.   Number of convenience stores </font>

<font color='yellow'>**STEP 1 - Data Collection **

---

We will used python panda library to load data (collected from Kaggle)</font>


```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Python/AI/datasets/Real estate.csv")
df.head()
```

<font color='yellow'>**STEP 2 - Data Cleaning **

---
Logically Colums like -
1.   transaction date
2.   latitude
3.   longitude

Should not have any influence on the outcome (price per unit). Hence, we should remove them </font>




```python
df=df.drop(['No','X1 transaction date', 'X5 latitude','X6 longitude'], axis=1)
```

<font color='yellow'>**STEP 3 - Data Preparation **

---
<font color='yellow'>In this step we will select input and result data and create train and test data sample.

<font color='yellow'>In current example we will consider

input x -> age of the house,
distance from nearest MRT satiation,
number of convenience stores</font>

ouput y ->house price per unit area
</font>

```python
x=df.iloc[:,0:3]
y=df.iloc[:,3:4]
```

<font color='yellow'>Using Python library for creating train and test split .2 = 80-20% split

```python
from sklearn import model_selection
X_train,X_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2,random_state=2)
```

<font color='yellow'>**STEP 4 - Model Selection **

---
<font color='yellow'>Process of fitting the dataset into a standard ML model algorithm . Our current dataset is best fit for Linear Regression algorithm. We will use standard Python sklearn "LinearRegression" ML model algorithm </font>

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

<font color='yellow'>**STEP 5 - Model Training **

---
We will used standard libraray method to train our model with train data set created in step-3</font>

```python
model.fit(X_train,y_train)
```

<font color='yellow'>**STEP 6 - Model Evaluation **

---
We will used sample test data created in step-3 to evaluate the model performance. Model's standard methods will be used.</font>

```python
y_predict=model.predict(X_test)
```

<font color='yellow'>Depending on the use-case and type of the model selection various performance matrices will be collected to evaluate the accuracy of the model. (standard libraried will be used)</font>

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
MAE = mean_absolute_error(y_test,y_predict)
MSE = mean_squared_error(y_test,y_predict)
R2_score = r2_score(y_test,y_predict)
print('MAE -> ',MAE)
print('MSE -> ',MSE)
print('R2 score -> ',R2_score)
```

```python
model.predict([[16,100,7]])
```

<font color='yellow'>**STEP 7 - Model deployment **

---
Trained model will be exported. This will be used in web framework to host the model as API</font>

```python
import pickle
with open('/content/drive/MyDrive/Python/AI/models/model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
